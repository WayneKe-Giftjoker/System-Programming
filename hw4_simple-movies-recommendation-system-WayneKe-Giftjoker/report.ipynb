```python
import matplotlib.pyplot as plt
import statistics
import matplotlib.ticker as ticker
from  matplotlib.ticker  import  MaxNLocator

layer_num = [1, 2, 3, 4, 5]
pserver_t = [[17.862, 14.121, 11.990, 13.810, 16.179], [3.509, 3.403, 3.664, 3.377, 3.733], [1.165, 1.257, 1.211, 1.251, 1.400], [0.702, 0.685, 0.737, 0.820, 0.686], [0.584, 0.685, 0.590, 0.575, 0.584]]
tserver_t = [[14.859, 14.157, 17.265, 15.820, 17.017], [3.656, 3.645, 3.651, 3.609, 3.321], [1.238, 1.405, 1.213, 1.290, 1.176], [0.611, 0.759, 0.675, 0.671, 0.653], [0.615, 0.649, 0.639, 0.647, 0.712]]
```

```python
pserver_avg_t = []
tserver_avg_t = []

for i in range(5):
    pserver_avg_t.append(round(statistics.mean(pserver_t[i]), 3))
    tserver_avg_t.append(round(statistics.mean(tserver_t[i]), 3))
print(pserver_avg_t)
print(tserver_avg_t)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.title("Real Execution Time", fontsize=35) # title
plt.ylabel("Time (sec)", fontsize=20) # y label
plt.xlabel("Layer Number", fontsize=20) # x label



plt.plot(layer_num, pserver_avg_t, color=(255/255,100/255,100/255), label= 'pserver', marker = ".")
plt.plot(layer_num, tserver_avg_t, '--', color=(100/255,100/255,255/255), label = 'tserver', marker = ".")

plt.gca().xaxis.set_major_locator(MaxNLocator( integer = True ))
# plt.gca().yaxis.set_major_locator(MaxNLocator( float = True ))


# lines = plt.plot(layer_num, pserver_avg_t, layer_num, tserver_avg_t)
# plt.setp(lines, marker = "o")
plt.grid(True)
# plt.figure(figsize=(6, 8))
plt.legend(loc = 'upper right', shadow = True, fontsize=15)

for index in range(len(layer_num)):
      ax.text(layer_num[index], pserver_avg_t[index], '({:.0f}, {:.3f}s)'.format(layer_num[index],pserver_avg_t[index]), size=12, color=(255/255,100/255,100/255), va = "top", ha="right")
      ax.text(layer_num[index], tserver_avg_t[index], '({:.0f}, {:.3f}s)'.format(layer_num[index],tserver_avg_t[index]), size=12, color=(100/255,100/255,255/255), va = "bottom", ha = "left")

# fig , ax = plt.subplots()
fig.set_size_inches(15, 10)
# ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.3f'))
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.3f'))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter(useMathText=True))
# leg = ax.legend(loc='upper right', shadow=True)


plt.show()
fig.savefig("Real Execution Time.jpg");
```
